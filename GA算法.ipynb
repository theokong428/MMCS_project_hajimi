```python
import pandas as pd
import numpy as np
import xpress as xp
xp.init('/Applications/FICO Xpress/xpressmp/bin/xpauth.xpr')

from haversine import haversine, Unit
import warnings
from geneticalgorithm import geneticalgorithm as ga
import sys
import time

# 暂时忽略 Pandas 的警告

print("--- 阶段 1: 数据准备与索引映射 ---")

# --- 1.1 加载站点数据 (库存与容量) 并强制站点ID为字符串 ---
try:
    df_capacity = pd.read_csv("Nomin_capacity_daytype_1.csv")
    df_capacity['station_id'] = df_capacity['station_id'].astype(str)
    
    # 转换为字典
    initial_stock_raw = df_capacity.set_index('station_id')['x_i'].to_dict()
    capacity_raw = df_capacity.set_index('station_id')['y_i'].to_dict()
    STATIONS_RAW = list(initial_stock_raw.keys()) # 所有在容量文件中的站点
except Exception as e:
    print(f"Error loading result_capacity.csv: {e}")
    sys.exit(1)

# --- 1.2 加载经纬度数据并统一站点 ID 格式 ---
try:
    df_coords = pd.read_csv("merged_2018_2021.csv")
    df_coords['start_station_id'] = df_coords['start_station_id'].astype(str)
    df_coords['end_station_id'] = df_coords['end_station_id'].astype(str)
    
    station_coords_raw = {}
    
    # A. 从起始站点提取坐标
    df_start_coords = df_coords[['start_station_id', 'start_station_latitude', 'start_station_longitude']].dropna().drop_duplicates(subset=['start_station_id'])
    for row in df_start_coords.itertuples():
        station_coords_raw[row.start_station_id] = {
            'lat': float(row.start_station_latitude),
            'lon': float(row.start_station_longitude)
        }

    # B. 补充从结束站点提取的坐标
    df_end_coords = df_coords[['end_station_id', 'end_station_latitude', 'end_station_longitude']].dropna().drop_duplicates(subset=['end_station_id'])
    for row in df_end_coords.itertuples():
        station_id = row.end_station_id
        if station_id not in station_coords_raw:
            station_coords_raw[station_id] = {
                'lat': float(row.end_station_latitude), 
                'lon': float(row.end_station_longitude)
            }
except Exception as e:
    print(f"Error loading merged_2018_2021.csv: {e}")
    sys.exit(1)

# --- 1.3 关键数据过滤：只保留同时具有容量和坐标的站点 ---
STATIONS = [sid for sid in STATIONS_RAW if sid in station_coords_raw]
missing_stations = set(STATIONS_RAW) - set(STATIONS)
if missing_stations:
    print(f"注意: {len(missing_stations)} 个站点因缺少坐标被排除。")

# --- 1.4 新增：创建索引映射 ---
# 1. 创建映射字典
# {'1017': 0, '1018': 1, '1020': 2, ...}
id_to_int = {station_id: i for i, station_id in enumerate(STATIONS)}
# {0: '1017', 1: '1018', 2: '1020', ...}
int_to_id = {i: station_id for i, station_id in enumerate(STATIONS)}
STATIONS_INT = list(int_to_id.keys()) # [0, 1, 2, ...]
N_STATIONS = len(STATIONS_INT)

# 2. 将所有数据字典的键转换为整数索引
initial_stock = {id_to_int[sid]: val for sid, val in initial_stock_raw.items() if sid in id_to_int}
capacity = {id_to_int[sid]: val for sid, val in capacity_raw.items() if sid in id_to_int}

# --- 1.5 模拟需求 (必须替换为您的真实预测) ---
# (模拟) 需求预测 (d^B_i, d^R_i) - 已映射为整数索引
np.random.seed(42)
demand_B = {i: np.random.randint(5, 15) for i in STATIONS_INT}
demand_R = {i: np.random.randint(5, 15) for i in STATIONS_INT}

# --- 1.6 计算整数索引的距离矩阵 ---
distance_matrix = pd.DataFrame(index=STATIONS_INT, columns=STATIONS_INT, dtype=float)
for i in STATIONS_INT:
    for j in STATIONS_INT:
        if i == j:
            distance_matrix.loc[i, j] = 0.0
            continue
        # 解码回原始 ID
        sid_i = int_to_id[i] 
        sid_j = int_to_id[j]
        # 查找坐标
        coord_i = (station_coords_raw[sid_i]['lat'], station_coords_raw[sid_i]['lon'])
        coord_j = (station_coords_raw[sid_j]['lat'], station_coords_raw[sid_j]['lon'])
        # 计算距离
        distance_matrix.loc[i, j] = haversine(coord_i, coord_j, unit=Unit.KILOMETERS)

# 保存距离矩阵
output_filename = 'distance_matrix_int_km.csv'
distance_matrix.to_csv(output_filename)
print(f"整数索引的距离矩阵已保存到 {output_filename}")

# --- 1.7 定义成本和车辆参数 ---
Cost_Unmet_Borrow = 50       
Cost_Unmet_Return = 50       
Cost_Movement_KM = 0.5       
Cost_Movement_Unit_KM = 0.1  
Vehicle_Capacity_Dict = {1: 20, 2: 20, 3: 20, 4: 20, 5: 20, 6: 20, 7: 20, 8: 20, 9: 20, 10: 20} # K=1, K=2
VEHICLES = list(Vehicle_Capacity_Dict.keys())
M_big = 10000 # Xpress 中使用的大 M

print(f"数据准备完毕。共 {N_STATIONS} 个有效站点。")


# --- 阶段 2: Xpress 子问题 (适应度函数) ---

def solve_subproblem(paths_by_vehicle_int, depot_int, K):
    """
    Xpress 求解器: 计算给定整数索引路径下的最优流量和最小惩罚成本
    """
    model = xp.problem(name='IRP_Subproblem')
    model.setControl('outputlog', 0) # 关闭 Xpress 求解日志

    # 1. 从 GA 路径中提取 V_ijk (固定参数)
    V_ijk = {} # (i, j, k) -> 1
    u_k = {}   # k -> 1 (车辆使用状态)
    cost_vrp_travel = 0.0 # VRP 路径成本 (3.17)
    
    for k, path in paths_by_vehicle_int.items():
        if k > K: continue # 确保不超过车辆数
        if len(path) > 2:
            u_k[k] = 1
            for i in range(len(path) - 1):
                start_node = path[i]
                end_node = path[i+1]
                
                # 检查距离 (理论上不应发生，因为已过滤)
                if np.isinf(distance_matrix.loc[start_node, end_node]):
                    return 0.00001
                    
                V_ijk[start_node, end_node, k] = 1
                cost_vrp_travel += distance_matrix.loc[start_node, end_node] * Cost_Movement_KM
        else:
            u_k[k] = 0

    # 2. 定义 Xpress 决策变量 (使用整数索引)
    x = {} # x_ijk (流量, 只能在 V_ijk=1 的路径上流动)
    for (i, j, k) in V_ijk.keys():
        x[i, j, k] = xp.var(name=f'x_{i}_{j}_{k}', vartype=xp.integer, lb=0)
    
    q = {i: {k: xp.var(name=f'q_{i}_{k}', lb=0) for k in VEHICLES} for i in STATIONS_INT}
    z_B = {i: xp.var(name=f'z_B_{i}', lb=0) for i in STATIONS_INT}
    z_R = {i: xp.var(name=f'z_R_{i}', lb=0) for i in STATIONS_INT}
    s_end = {i: xp.var(name=f's_end_{i}', lb=0) for i in STATIONS_INT}
    
    model.addVariable(list(x.values()), [q[i][k] for i in STATIONS_INT for k in VEHICLES], 
                      list(z_B.values()), list(z_R.values()), list(s_end.values()))

    # 3. 定义子问题目标函数 (最小化流量运输成本 + 惩罚成本)
    cost_flow = xp.Sum(distance_matrix.loc[i, j] * Cost_Movement_Unit_KM * x[i, j, k] 
                       for (i, j, k) in x.keys())
    
    cost_unmet = xp.Sum(Cost_Unmet_Borrow * z_B[i] + Cost_Unmet_Return * z_R[i] 
                        for i in STATIONS_INT)
    
    model.setObjective(cost_flow + cost_unmet, sense=xp.minimize)

    # 4. 添加约束 (使用整数索引字典)
    for i in STATIONS_INT:
        # (3.3) 库存平衡
        flow_in = xp.Sum(x.get((j, i, k), 0) for j in STATIONS_INT for k in VEHICLES if j != i)
        flow_out = xp.Sum(x.get((i, j, k), 0) for j in STATIONS_INT for k in VEHICLES if i != j)
        model.addConstraint(
            initial_stock.get(i, 0) + flow_in - flow_out + demand_R.get(i, 0) - demand_B.get(i, 0) - z_R[i] + z_B[i] == s_end[i]
        )
        # (3.7) (3.8) 容量约束
        model.addConstraint(s_end[i] <= capacity.get(i, M_big))
        # (3.9) (3.10) 未满足需求约束
        model.addConstraint(z_B[i] <= demand_B.get(i, 0))
        model.addConstraint(z_R[i] <= demand_R.get(i, 0))

    # (3.11) - (3.14) 车辆载重约束
    for k in VEHICLES:
        if u_k.get(k, 0) == 0: continue
        
        for i in STATIONS_INT:
            model.addConstraint(q[i][k] <= Vehicle_Capacity_Dict.get(k, M_big))
            
            p_ik = xp.Sum(x.get((i, j, k), 0) for j in STATIONS_INT if i != j) # 从 i 卸货
            d_ik = xp.Sum(x.get((j, i, k), 0) for j in STATIONS_INT if j != i) # 从 i 装货
            
            # (3.11) 载重平衡 (流平衡)
            previous_nodes = [j for j in STATIONS_INT if V_ijk.get((j, i, k), 0) == 1]
            if i != depot_int and previous_nodes:
                j = previous_nodes[0] # 假设单向路径
                model.addConstraint(q[i][k] == q[j][k] + d_ik - p_ik)

        # (3.12) (3.13) 车辆从 Depot 出发时为空
        if depot_int in STATIONS_INT:
             model.addConstraint(q[depot_int][k] == 0)

        # 5. 求解子问题
    try:
        model.solve()
    except Exception:
        return 0.00001  # 求解失败

    # 获取求解状态字符串（最稳定方法）
    status_str = model.getProbStatusString()

    # 判断是否最优
    if "optimal" in status_str.lower():
        cost_subproblem = model.getObjVal()
        total_cost = cost_vrp_travel + cost_subproblem
        return 1.0 / total_cost   # 适应度（最大化）
    else:
        return 0.00001  # 不可行解

print("--- 阶段 2: Xpress 子问题 (适应度函数) 定义完毕 ---")


# --- 阶段 3: GA 主循环 (索引映射 + 随机键编码) ---

# 1. 定义 GA 算法的参数
ga_parameters = {
    'max_num_iteration': 100,         
    'population_size': 50,            
    'parents_portion': 0.3,           
    'mutation_probability': 0.1,      
    'crossover_probability': 0.9,     
    'elit_ratio': 0.01,               
    'crossover_type': 'uniform',      
    'selection_type': 'roulette',     
    'max_iteration_without_improv': None 
}

# 2. 外循环：求解 K 和 Depot
# 假设 '1017' 是 Depot 候选站
candidate_depots_str = ['1017'] 
candidate_depots_int = [id_to_int[s] for s in candidate_depots_str if s in id_to_int]

if not candidate_depots_int:
     print("警告: 候选 Depot 站点 '1017' 缺失坐标，正在使用第一个可用站点作为 Depot。")
     candidate_depots_int = [STATIONS_INT[0]] if STATIONS_INT else [] 

if not STATIONS_INT:
    print("错误: 没有可用的站点进行优化。")
    sys.exit(1)

max_vehicles = 3 # 假设最多 10 辆车
best_overall_cost = float('inf')
best_solution = {}

start_time = time.time()

for depot_int in candidate_depots_int:
    
    # 待访问站点 = [0, 1, 2...] (不包括 depot_int)
    stations_to_visit_int = [i for i in STATIONS_INT if i != depot_int]

    for k in range(1, max_vehicles + 1):
        
        print(f"--- 正在运行 GA: Depot={int_to_id[depot_int]}, K={k} 辆车 ---")
        
        # 基因池 = [0, 1, 2...] (待访问) + [depot_int] * (k-1) (分隔符)
        genes_to_permute = stations_to_visit_int + [depot_int] * (k - 1)
        dimension = len(genes_to_permute)
        
        if dimension == 0:
            continue

        # 适应度函数封装 (使用 Random-Key 解码)
        def fitness_wrapper(chromosome):
            # 1. 解码：chromosome 是 [0,1] 间的浮点数
            permutation_indices = np.argsort(chromosome)
            
            # 2. 映射：从基因池中构建有序的整数 ID 路径
            ordered_path_list = [genes_to_permute[i] for i in permutation_indices]
            
            # 3. 解析路径 (整数索引)
            paths_by_vehicle_int = {}
            current_path = [depot_int]
            current_k = 1
            
            for station_id_int in ordered_path_list:
                if station_id_int == depot_int: # 遇到分隔符 (Depot)
                    current_path.append(depot_int)
                    paths_by_vehicle_int[current_k] = current_path
                    current_k += 1
                    current_path = [depot_int]
                else:
                    current_path.append(station_id_int)
            
            current_path.append(depot_int)
            paths_by_vehicle_int[current_k] = current_path
            
            # 4. 调用 Xpress 子问题 (传入整数路径)
            return solve_subproblem(paths_by_vehicle_int, depot_int, k)



        # 增加 Xpress 求解的超时时间，防止 GA 强制中断
        # 建议设置为 None 禁用 GA 库的超时，让 Xpress 控制求解时间
        TIMEOUT_SECONDS = 3600
        varbound = np.array([[0.0, 1.0]] * dimension)
        ga_model = ga(function=fitness_wrapper,
                      dimension=dimension,
                      variable_type='real',
                      variable_boundaries=varbound,
                      algorithm_parameters=ga_parameters,
                      function_timeout=TIMEOUT_SECONDS) # 禁用 GA 库的默认 10.0 秒超时

         

        ga_model.run()
        
        # 提取结果
        if ga_model.best_function > 0.00001:
            current_cost = 1.0 / ga_model.best_function
        else:
            current_cost = float('inf')
            
        print(f"    K={k} 完成。最优成本: {current_cost:.2f}")

        if current_cost < best_overall_cost:
            best_overall_cost = current_cost
            best_solution = {
                'Depot_Int': depot_int,
                'K_Vehicles': k,
                'Total_Cost': current_cost,
                'Optimal_Path_Chromosome (Float-Key)': ga_model.best_variable
            }

# --- 阶段 4: 结果解码与输出 ---
end_time = time.time()
print(f"\n--- 求解完成 (总耗时: {end_time - start_time:.2f} 秒) ---")

if not best_solution:
    print("未找到可行的求解方案。")
else:
    print(f"调度中心 (Depot): {int_to_id.get(best_solution.get('Depot_Int'), 'N/A')}")
    print(f"调度车辆数量 (K): {best_solution.get('K_Vehicles')}")
    print(f"最小总成本: {best_solution.get('Total_Cost'):.2f}")
    
    # 补充：最终路径解码
    final_chromosome = best_solution.get('Optimal_Path_Chromosome (Float-Key)')
    depot_int = best_solution.get('Depot_Int')
    k = best_solution.get('K_Vehicles')
    
    if final_chromosome is not None and depot_int is not None:
        stations_to_visit_int = [i for i in STATIONS_INT if i != depot_int]
        genes_to_permute = stations_to_visit_int + [depot_int] * (k - 1)
        
        permutation_indices = np.argsort(final_chromosome)
        ordered_path_list = [genes_to_permute[i] for i in permutation_indices]

        final_paths = {}
        current_path = [depot_int]
        current_k = 1
        for station_id_int in ordered_path_list:
            if station_id_int == depot_int:
                current_path.append(depot_int)
                final_paths[current_k] = current_path
                current_k += 1
                current_path = [depot_int]
            else:
                current_path.append(station_id_int)
        current_path.append(depot_int)
        final_paths[current_k] = current_path
        
        print("\n--- 最终调度路径 (原始 ID) ---")
        for k_id, path_int in final_paths.items():
            if len(path_int) > 2:
                # 解码为原始 ID
                path_str = [int_to_id[i] for i in path_int]
                print(f"车辆 {k_id} 路径: {' -> '.join(path_str)}")
```

```python
import pandas as pd
import numpy as np
import xpress as xp
xp.init('/Applications/FICO Xpress/xpressmp/bin/xpauth.xpr')

from haversine import haversine, Unit
import warnings
from geneticalgorithm import geneticalgorithm as ga
import sys
import time

# 暂时忽略 Pandas 的警告

print("--- 阶段 1: 数据准备与索引映射 ---")

# --- 1.1 加载站点数据 (库存与容量) 并强制站点ID为字符串 ---
try:
    df_capacity = pd.read_csv("Nomin_capacity_daytype_1.csv")
    df_capacity['station_id'] = df_capacity['station_id'].astype(str)
    
    # 转换为字典
    initial_stock_raw = df_capacity.set_index('station_id')['x_i'].to_dict()
    capacity_raw = df_capacity.set_index('station_id')['y_i'].to_dict()
    STATIONS_RAW = list(initial_stock_raw.keys()) # 所有在容量文件中的站点
except Exception as e:
    print(f"Error loading result_capacity.csv: {e}")
    sys.exit(1)

# --- 1.2 加载经纬度数据并统一站点 ID 格式 ---
try:
    df_coords = pd.read_csv("merged_2018_2021.csv")
    df_coords['start_station_id'] = df_coords['start_station_id'].astype(str)
    df_coords['end_station_id'] = df_coords['end_station_id'].astype(str)
    
    station_coords_raw = {}
    
    # A. 从起始站点提取坐标
    df_start_coords = df_coords[['start_station_id', 'start_station_latitude', 'start_station_longitude']].dropna().drop_duplicates(subset=['start_station_id'])
    for row in df_start_coords.itertuples():
        station_coords_raw[row.start_station_id] = {
            'lat': float(row.start_station_latitude),
            'lon': float(row.start_station_longitude)
        }

    # B. 补充从结束站点提取的坐标
    df_end_coords = df_coords[['end_station_id', 'end_station_latitude', 'end_station_longitude']].dropna().drop_duplicates(subset=['end_station_id'])
    for row in df_end_coords.itertuples():
        station_id = row.end_station_id
        if station_id not in station_coords_raw:
            station_coords_raw[station_id] = {
                'lat': float(row.end_station_latitude), 
                'lon': float(row.end_station_longitude)
            }
except Exception as e:
    print(f"Error loading merged_2018_2021.csv: {e}")
    sys.exit(1)

# --- 1.3 关键数据过滤：只保留同时具有容量和坐标的站点 ---
STATIONS = [sid for sid in STATIONS_RAW if sid in station_coords_raw]
missing_stations = set(STATIONS_RAW) - set(STATIONS)
if missing_stations:
    print(f"注意: {len(missing_stations)} 个站点因缺少坐标被排除。")

# --- 1.4 新增：创建索引映射 ---
# 1. 创建映射字典
# {'1017': 0, '1018': 1, '1020': 2, ...}
id_to_int = {station_id: i for i, station_id in enumerate(STATIONS)}
# {0: '1017', 1: '1018', 2: '1020', ...}
int_to_id = {i: station_id for i, station_id in enumerate(STATIONS)}
STATIONS_INT = list(int_to_id.keys()) # [0, 1, 2, ...]
N_STATIONS = len(STATIONS_INT)

# 2. 将所有数据字典的键转换为整数索引
initial_stock = {id_to_int[sid]: val for sid, val in initial_stock_raw.items() if sid in id_to_int}
capacity = {id_to_int[sid]: val for sid, val in capacity_raw.items() if sid in id_to_int}

# --- 1.5 模拟需求 (必须替换为您的真实预测) ---
# (模拟) 需求预测 (d^B_i, d^R_i) - 已映射为整数索引
np.random.seed(42)
demand_B = {i: np.random.randint(5, 15) for i in STATIONS_INT}
demand_R = {i: np.random.randint(5, 15) for i in STATIONS_INT}

# --- 1.6 计算整数索引的距离矩阵 ---
distance_matrix = pd.DataFrame(index=STATIONS_INT, columns=STATIONS_INT, dtype=float)
for i in STATIONS_INT:
    for j in STATIONS_INT:
        if i == j:
            distance_matrix.loc[i, j] = 0.0
            continue
        # 解码回原始 ID
        sid_i = int_to_id[i] 
        sid_j = int_to_id[j]
        # 查找坐标
        coord_i = (station_coords_raw[sid_i]['lat'], station_coords_raw[sid_i]['lon'])
        coord_j = (station_coords_raw[sid_j]['lat'], station_coords_raw[sid_j]['lon'])
        # 计算距离
        distance_matrix.loc[i, j] = haversine(coord_i, coord_j, unit=Unit.KILOMETERS)

# 保存距离矩阵
output_filename = 'distance_matrix_int_km.csv'
distance_matrix.to_csv(output_filename)
print(f"整数索引的距离矩阵已保存到 {output_filename}")

# --- 1.7 定义成本和车辆参数 ---
Cost_Unmet_Borrow = 50       
Cost_Unmet_Return = 50       
Cost_Movement_KM = 0.5       
Cost_Movement_Unit_KM = 0.1  
Vehicle_Capacity_Dict = {1: 20, 2: 20, 3: 20, 4: 20, 5: 20, 6: 20, 7: 20, 8: 20, 9: 20, 10: 20} # K=1, K=2
VEHICLES = list(Vehicle_Capacity_Dict.keys())
M_big = 10000 # Xpress 中使用的大 M

print(f"数据准备完毕。共 {N_STATIONS} 个有效站点。")


# --- 阶段 2: Xpress 子问题 (适应度函数) ---

def solve_subproblem(paths_by_vehicle_int, depot_int, K):
    """
    Xpress 求解器: 计算给定整数索引路径下的最优流量和最小惩罚成本
    """
    model = xp.problem(name='IRP_Subproblem')
    model.setControl('outputlog', 0) # 关闭 Xpress 求解日志

    # 1. 从 GA 路径中提取 V_ijk (固定参数)
    V_ijk = {} # (i, j, k) -> 1
    u_k = {}   # k -> 1 (车辆使用状态)
    cost_vrp_travel = 0.0 # VRP 路径成本 (3.17)
    
    for k, path in paths_by_vehicle_int.items():
        if k > K: continue # 确保不超过车辆数
        if len(path) > 2:
            u_k[k] = 1
            for i in range(len(path) - 1):
                start_node = path[i]
                end_node = path[i+1]
                
                # 检查距离 (理论上不应发生，因为已过滤)
                if np.isinf(distance_matrix.loc[start_node, end_node]):
                    return 0.00001
                    
                V_ijk[start_node, end_node, k] = 1
                cost_vrp_travel += distance_matrix.loc[start_node, end_node] * Cost_Movement_KM
        else:
            u_k[k] = 0

    # 2. 定义 Xpress 决策变量 (使用整数索引)
    x = {} # x_ijk (流量, 只能在 V_ijk=1 的路径上流动)
    for (i, j, k) in V_ijk.keys():
        x[i, j, k] = xp.var(name=f'x_{i}_{j}_{k}', vartype=xp.integer, lb=0)
    
    q = {i: {k: xp.var(name=f'q_{i}_{k}', lb=0) for k in VEHICLES} for i in STATIONS_INT}
    z_B = {i: xp.var(name=f'z_B_{i}', lb=0) for i in STATIONS_INT}
    z_R = {i: xp.var(name=f'z_R_{i}', lb=0) for i in STATIONS_INT}
    s_end = {i: xp.var(name=f's_end_{i}', lb=0) for i in STATIONS_INT}
    
    model.addVariable(list(x.values()), [q[i][k] for i in STATIONS_INT for k in VEHICLES], 
                      list(z_B.values()), list(z_R.values()), list(s_end.values()))

    # 3. 定义子问题目标函数 (最小化流量运输成本 + 惩罚成本)
    cost_flow = xp.Sum(distance_matrix.loc[i, j] * Cost_Movement_Unit_KM * x[i, j, k] 
                       for (i, j, k) in x.keys())
    
    cost_unmet = xp.Sum(Cost_Unmet_Borrow * z_B[i] + Cost_Unmet_Return * z_R[i] 
                        for i in STATIONS_INT)
    
    model.setObjective(cost_flow + cost_unmet, sense=xp.minimize)

    # 4. 添加约束 (使用整数索引字典)
    for i in STATIONS_INT:
        # (3.3) 库存平衡
        flow_in = xp.Sum(x.get((j, i, k), 0) for j in STATIONS_INT for k in VEHICLES if j != i)
        flow_out = xp.Sum(x.get((i, j, k), 0) for j in STATIONS_INT for k in VEHICLES if i != j)
        model.addConstraint(
            initial_stock.get(i, 0) + flow_in - flow_out + demand_R.get(i, 0) - demand_B.get(i, 0) - z_R[i] + z_B[i] == s_end[i]
        )
        # (3.7) (3.8) 容量约束
        model.addConstraint(s_end[i] <= capacity.get(i, M_big))
        # (3.9) (3.10) 未满足需求约束
        model.addConstraint(z_B[i] <= demand_B.get(i, 0))
        model.addConstraint(z_R[i] <= demand_R.get(i, 0))

    # (3.11) - (3.14) 车辆载重约束
    for k in VEHICLES:
        if u_k.get(k, 0) == 0: continue
        
        for i in STATIONS_INT:
            model.addConstraint(q[i][k] <= Vehicle_Capacity_Dict.get(k, M_big))
            
            p_ik = xp.Sum(x.get((i, j, k), 0) for j in STATIONS_INT if i != j) # 从 i 卸货
            d_ik = xp.Sum(x.get((j, i, k), 0) for j in STATIONS_INT if j != i) # 从 i 装货
            
            # (3.11) 载重平衡 (流平衡)
            previous_nodes = [j for j in STATIONS_INT if V_ijk.get((j, i, k), 0) == 1]
            if i != depot_int and previous_nodes:
                j = previous_nodes[0] # 假设单向路径
                model.addConstraint(q[i][k] == q[j][k] + d_ik - p_ik)

        # (3.12) (3.13) 车辆从 Depot 出发时为空
        if depot_int in STATIONS_INT:
             model.addConstraint(q[depot_int][k] == 0)

        # 5. 求解子问题
    try:
        model.solve()
    except Exception:
        return 0.00001  # 求解失败

    # 获取求解状态字符串（最稳定方法）
    status_str = model.getProbStatusString()

    # 判断是否最优
    if "optimal" in status_str.lower():
        cost_subproblem = model.getObjVal()
        total_cost = cost_vrp_travel + cost_subproblem
        return 1.0 / total_cost   # 适应度（最大化）
    else:
        return 0.00001  # 不可行解

print("--- 阶段 2: Xpress 子问题 (适应度函数) 定义完毕 ---")


# --- 阶段 3: GA 主循环 (索引映射 + 随机键编码) ---

# 1. 定义 GA 算法的参数
ga_parameters = {
    'max_num_iteration': 100,         
    'population_size': 50,            
    'parents_portion': 0.3,           
    'mutation_probability': 0.1,      
    'crossover_probability': 0.9,     
    'elit_ratio': 0.01,               
    'crossover_type': 'uniform',      
    'selection_type': 'roulette',     
    'max_iteration_without_improv': None 
}

# 2. 外循环：求解 K 和 Depot
# 假设 '1017' 是 Depot 候选站
candidate_depots_str = STATIONS_INT 
candidate_depots_int = [id_to_int[s] for s in candidate_depots_str if s in id_to_int]


if not STATIONS_INT:
    print("错误: 没有可用的站点进行优化。")
    sys.exit(1)

max_vehicles = 3 # 假设最多 10 辆车
best_overall_cost = float('inf')
best_solution = {}

start_time = time.time()

for depot_int in candidate_depots_int:
    
    # 待访问站点 = [0, 1, 2...] (不包括 depot_int)
    stations_to_visit_int = [i for i in STATIONS_INT if i != depot_int]

    for k in range(1, max_vehicles + 1):
        
        print(f"--- 正在运行 GA: Depot={int_to_id[depot_int]}, K={k} 辆车 ---")
        
        # 基因池 = [0, 1, 2...] (待访问) + [depot_int] * (k-1) (分隔符)
        genes_to_permute = stations_to_visit_int + [depot_int] * (k - 1)
        dimension = len(genes_to_permute)
        
        if dimension == 0:
            continue

        # 适应度函数封装 (使用 Random-Key 解码)
        def fitness_wrapper(chromosome):
            # 1. 解码：chromosome 是 [0,1] 间的浮点数
            permutation_indices = np.argsort(chromosome)
            
            # 2. 映射：从基因池中构建有序的整数 ID 路径
            ordered_path_list = [genes_to_permute[i] for i in permutation_indices]
            
            # 3. 解析路径 (整数索引)
            paths_by_vehicle_int = {}
            current_path = [depot_int]
            current_k = 1
            
            for station_id_int in ordered_path_list:
                if station_id_int == depot_int: # 遇到分隔符 (Depot)
                    current_path.append(depot_int)
                    paths_by_vehicle_int[current_k] = current_path
                    current_k += 1
                    current_path = [depot_int]
                else:
                    current_path.append(station_id_int)
            
            current_path.append(depot_int)
            paths_by_vehicle_int[current_k] = current_path
            
            # 4. 调用 Xpress 子问题 (传入整数路径)
            return solve_subproblem(paths_by_vehicle_int, depot_int, k)



        # 增加 Xpress 求解的超时时间，防止 GA 强制中断
        # 建议设置为 None 禁用 GA 库的超时，让 Xpress 控制求解时间
        TIMEOUT_SECONDS = 3600
        varbound = np.array([[0.0, 1.0]] * dimension)
        ga_model = ga(function=fitness_wrapper,
                      dimension=dimension,
                      variable_type='real',
                      variable_boundaries=varbound,
                      algorithm_parameters=ga_parameters,
                      function_timeout=TIMEOUT_SECONDS) # 禁用 GA 库的默认 10.0 秒超时

         

        ga_model.run()
        
        # 提取结果
        if ga_model.best_function > 0.00001:
            current_cost = 1.0 / ga_model.best_function
        else:
            current_cost = float('inf')
            
        print(f"    K={k} 完成。最优成本: {current_cost:.2f}")

        if current_cost < best_overall_cost:
            best_overall_cost = current_cost
            best_solution = {
                'Depot_Int': depot_int,
                'K_Vehicles': k,
                'Total_Cost': current_cost,
                'Optimal_Path_Chromosome (Float-Key)': ga_model.best_variable
            }

# --- 阶段 4: 结果解码与输出 ---
end_time = time.time()
print(f"\n--- 求解完成 (总耗时: {end_time - start_time:.2f} 秒) ---")

if not best_solution:
    print("未找到可行的求解方案。")
else:
    print(f"调度中心 (Depot): {int_to_id.get(best_solution.get('Depot_Int'), 'N/A')}")
    print(f"调度车辆数量 (K): {best_solution.get('K_Vehicles')}")
    print(f"最小总成本: {best_solution.get('Total_Cost'):.2f}")
    
    # 补充：最终路径解码
    final_chromosome = best_solution.get('Optimal_Path_Chromosome (Float-Key)')
    depot_int = best_solution.get('Depot_Int')
    k = best_solution.get('K_Vehicles')
    
    if final_chromosome is not None and depot_int is not None:
        stations_to_visit_int = [i for i in STATIONS_INT if i != depot_int]
        genes_to_permute = stations_to_visit_int + [depot_int] * (k - 1)
        
        permutation_indices = np.argsort(final_chromosome)
        ordered_path_list = [genes_to_permute[i] for i in permutation_indices]

        final_paths = {}
        current_path = [depot_int]
        current_k = 1
        for station_id_int in ordered_path_list:
            if station_id_int == depot_int:
                current_path.append(depot_int)
                final_paths[current_k] = current_path
                current_k += 1
                current_path = [depot_int]
            else:
                current_path.append(station_id_int)
        current_path.append(depot_int)
        final_paths[current_k] = current_path
        
        print("\n--- 最终调度路径 (原始 ID) ---")
        for k_id, path_int in final_paths.items():
            if len(path_int) > 2:
                # 解码为原始 ID
                path_str = [int_to_id[i] for i in path_int]
                print(f"车辆 {k_id} 路径: {' -> '.join(path_str)}")
```

```python
import pandas as pd
import numpy as np
import xpress as xp
xp.init('/Applications/FICO Xpress/xpressmp/bin/xpauth.xpr')

from haversine import haversine, Unit
import warnings
from geneticalgorithm import geneticalgorithm as ga
import sys
import time

# 暂时忽略 Pandas 的警告

print("--- 阶段 1: 数据准备与索引映射 ---")

# --- 1.1 加载站点数据 (库存与容量) 并强制站点ID为字符串 ---
try:
    df_capacity = pd.read_csv("Nomin_capacity_daytype_1.csv")
    df_capacity['station_id'] = df_capacity['station_id'].astype(str)
    
    # 转换为字典
    initial_stock_raw = df_capacity.set_index('station_id')['x_i'].to_dict()
    capacity_raw = df_capacity.set_index('station_id')['y_i'].to_dict()
    STATIONS_RAW = list(initial_stock_raw.keys()) # 所有在容量文件中的站点
except Exception as e:
    print(f"Error loading result_capacity.csv: {e}")
    sys.exit(1)

# --- 1.2 加载经纬度数据并统一站点 ID 格式 ---
try:
    df_coords = pd.read_csv("merged_2018_2021.csv")
    df_coords['start_station_id'] = df_coords['start_station_id'].astype(str)
    df_coords['end_station_id'] = df_coords['end_station_id'].astype(str)
    
    station_coords_raw = {}
    
    # A. 从起始站点提取坐标
    df_start_coords = df_coords[['start_station_id', 'start_station_latitude', 'start_station_longitude']].dropna().drop_duplicates(subset=['start_station_id'])
    for row in df_start_coords.itertuples():
        station_coords_raw[row.start_station_id] = {
            'lat': float(row.start_station_latitude),
            'lon': float(row.start_station_longitude)
        }

    # B. 补充从结束站点提取的坐标
    df_end_coords = df_coords[['end_station_id', 'end_station_latitude', 'end_station_longitude']].dropna().drop_duplicates(subset=['end_station_id'])
    for row in df_end_coords.itertuples():
        station_id = row.end_station_id
        if station_id not in station_coords_raw:
            station_coords_raw[station_id] = {
                'lat': float(row.end_station_latitude), 
                'lon': float(row.end_station_longitude)
            }
except Exception as e:
    print(f"Error loading merged_2018_2021.csv: {e}")
    sys.exit(1)

# --- 1.3 关键数据过滤：只保留同时具有容量和坐标的站点 ---
STATIONS = [sid for sid in STATIONS_RAW if sid in station_coords_raw]
missing_stations = set(STATIONS_RAW) - set(STATIONS)
if missing_stations:
    print(f"注意: {len(missing_stations)} 个站点因缺少坐标被排除。")

# --- 1.4 新增：创建索引映射 ---
# 1. 创建映射字典
# {'1017': 0, '1018': 1, '1020': 2, ...}
id_to_int = {station_id: i for i, station_id in enumerate(STATIONS)}
# {0: '1017', 1: '1018', 2: '1020', ...}
int_to_id = {i: station_id for i, station_id in enumerate(STATIONS)}
STATIONS_INT = list(int_to_id.keys()) # [0, 1, 2, ...]
N_STATIONS = len(STATIONS_INT)

# 2. 将所有数据字典的键转换为整数索引
initial_stock = {id_to_int[sid]: val for sid, val in initial_stock_raw.items() if sid in id_to_int}
capacity = {id_to_int[sid]: val for sid, val in capacity_raw.items() if sid in id_to_int}

# --- 1.5 模拟需求 (必须替换为您的真实预测) ---
# (模拟) 需求预测 (d^B_i, d^R_i) - 已映射为整数索引
np.random.seed(42)
demand_B = {i: np.random.randint(5, 15) for i in STATIONS_INT}
demand_R = {i: np.random.randint(5, 15) for i in STATIONS_INT}

# --- 1.6 计算整数索引的距离矩阵 ---
distance_matrix = pd.DataFrame(index=STATIONS_INT, columns=STATIONS_INT, dtype=float)
for i in STATIONS_INT:
    for j in STATIONS_INT:
        if i == j:
            distance_matrix.loc[i, j] = 0.0
            continue
        # 解码回原始 ID
        sid_i = int_to_id[i] 
        sid_j = int_to_id[j]
        # 查找坐标
        coord_i = (station_coords_raw[sid_i]['lat'], station_coords_raw[sid_i]['lon'])
        coord_j = (station_coords_raw[sid_j]['lat'], station_coords_raw[sid_j]['lon'])
        # 计算距离
        distance_matrix.loc[i, j] = haversine(coord_i, coord_j, unit=Unit.KILOMETERS)

# 保存距离矩阵
output_filename = 'distance_matrix_int_km.csv'
distance_matrix.to_csv(output_filename)
print(f"整数索引的距离矩阵已保存到 {output_filename}")

# --- 1.7 新增：计算最优 Depot 启发式 ---

print("正在计算最优 Depot 候选...")

best_depot_int = -1
min_total_distance = float('inf')

# 1. 遍历所有站点，将其作为候选 Depot
for candidate_depot_int in STATIONS_INT:
    
    # 计算该 Depot 到所有其他站点的距离总和
    # (我们不乘 2，只计算单程距离和，因为它只是一个排名指标)
    total_distance_sum = distance_matrix.loc[candidate_depot_int, :].sum()
    
    if total_distance_sum < min_total_distance:
        min_total_distance = total_distance_sum
        best_depot_int = candidate_depot_int

# 2. 将 GA 候选 Depot 列表更新为仅包含这个最优站点
if best_depot_int != -1:
    # 将整数索引解码回原始 ID
    best_depot_id = int_to_id[best_depot_int]
    
    print(f"启发式计算结果：最优 Depot 候选是 {best_depot_id} (索引: {best_depot_int})")
    
    # 将这个最优整数索引保存，用于下一阶段的 GA 循环
    OPTIMAL_CANDIDATE_DEPOTS_INT = [best_depot_int]
else:
    # 备用，如果出现意外错误
    OPTIMAL_CANDIDATE_DEPOTS_INT = STATIONS_INT[:1]



# --- 1.8 定义成本和车辆参数 ---
Cost_Unmet_Borrow = 50       
Cost_Unmet_Return = 50       
Cost_Movement_KM = 0.5       
Cost_Movement_Unit_KM = 0.1  
Vehicle_Capacity_Dict = {1: 20, 2: 20, 3: 20, 4: 20, 5: 20, 6: 20, 7: 20, 8: 20, 9: 20, 10: 20} # K=1, K=2
VEHICLES = list(Vehicle_Capacity_Dict.keys())
M_big = 10000 # Xpress 中使用的大 M

print(f"数据准备完毕。共 {N_STATIONS} 个有效站点。")


# --- 阶段 2: Xpress 子问题 (适应度函数) ---

def solve_subproblem(paths_by_vehicle_int, depot_int, K):
    """
    Xpress 求解器: 计算给定整数索引路径下的最优流量和最小惩罚成本
    """
    model = xp.problem(name='IRP_Subproblem')
    model.setControl('outputlog', 0) # 关闭 Xpress 求解日志

    # 1. 从 GA 路径中提取 V_ijk (固定参数)
    V_ijk = {} # (i, j, k) -> 1
    u_k = {}   # k -> 1 (车辆使用状态)
    cost_vrp_travel = 0.0 # VRP 路径成本 (3.17)
    
    for k, path in paths_by_vehicle_int.items():
        if k > K: continue # 确保不超过车辆数
        if len(path) > 2:
            u_k[k] = 1
            for i in range(len(path) - 1):
                start_node = path[i]
                end_node = path[i+1]
                
                # 检查距离 (理论上不应发生，因为已过滤)
                if np.isinf(distance_matrix.loc[start_node, end_node]):
                    return 0.00001
                    
                V_ijk[start_node, end_node, k] = 1
                cost_vrp_travel += distance_matrix.loc[start_node, end_node] * Cost_Movement_KM
        else:
            u_k[k] = 0

    # 2. 定义 Xpress 决策变量 (使用整数索引)
    x = {} # x_ijk (流量, 只能在 V_ijk=1 的路径上流动)
    for (i, j, k) in V_ijk.keys():
        x[i, j, k] = xp.var(name=f'x_{i}_{j}_{k}', vartype=xp.integer, lb=0)
    
    q = {i: {k: xp.var(name=f'q_{i}_{k}', lb=0) for k in VEHICLES} for i in STATIONS_INT}
    z_B = {i: xp.var(name=f'z_B_{i}', lb=0) for i in STATIONS_INT}
    z_R = {i: xp.var(name=f'z_R_{i}', lb=0) for i in STATIONS_INT}
    s_end = {i: xp.var(name=f's_end_{i}', lb=0) for i in STATIONS_INT}
    
    model.addVariable(list(x.values()), [q[i][k] for i in STATIONS_INT for k in VEHICLES], 
                      list(z_B.values()), list(z_R.values()), list(s_end.values()))

    # 3. 定义子问题目标函数 (最小化流量运输成本 + 惩罚成本)
    cost_flow = xp.Sum(distance_matrix.loc[i, j] * Cost_Movement_Unit_KM * x[i, j, k] 
                       for (i, j, k) in x.keys())
    
    cost_unmet = xp.Sum(Cost_Unmet_Borrow * z_B[i] + Cost_Unmet_Return * z_R[i] 
                        for i in STATIONS_INT)
    
    model.setObjective(cost_flow + cost_unmet, sense=xp.minimize)

    # 4. 添加约束 (使用整数索引字典)
    for i in STATIONS_INT:
        # (3.3) 库存平衡
        flow_in = xp.Sum(x.get((j, i, k), 0) for j in STATIONS_INT for k in VEHICLES if j != i)
        flow_out = xp.Sum(x.get((i, j, k), 0) for j in STATIONS_INT for k in VEHICLES if i != j)
        model.addConstraint(
            initial_stock.get(i, 0) + flow_in - flow_out + demand_R.get(i, 0) - demand_B.get(i, 0) - z_R[i] + z_B[i] == s_end[i]
        )
        # (3.7) (3.8) 容量约束
        model.addConstraint(s_end[i] <= capacity.get(i, M_big))
        # (3.9) (3.10) 未满足需求约束
        model.addConstraint(z_B[i] <= demand_B.get(i, 0))
        model.addConstraint(z_R[i] <= demand_R.get(i, 0))

    # (3.11) - (3.14) 车辆载重约束
    for k in VEHICLES:
        if u_k.get(k, 0) == 0: continue
        
        for i in STATIONS_INT:
            model.addConstraint(q[i][k] <= Vehicle_Capacity_Dict.get(k, M_big))
            
            p_ik = xp.Sum(x.get((i, j, k), 0) for j in STATIONS_INT if i != j) # 从 i 卸货
            d_ik = xp.Sum(x.get((j, i, k), 0) for j in STATIONS_INT if j != i) # 从 i 装货
            
            # (3.11) 载重平衡 (流平衡)
            previous_nodes = [j for j in STATIONS_INT if V_ijk.get((j, i, k), 0) == 1]
            if i != depot_int and previous_nodes:
                j = previous_nodes[0] # 假设单向路径
                model.addConstraint(q[i][k] == q[j][k] + d_ik - p_ik)

        # (3.12) (3.13) 车辆从 Depot 出发时为空
        if depot_int in STATIONS_INT:
             model.addConstraint(q[depot_int][k] == 0)

        # 5. 求解子问题
    try:
        model.solve()
    except Exception:
        return 0.00001  # 求解失败

    # 获取求解状态字符串（最稳定方法）
    status_str = model.getProbStatusString()

    # 判断是否最优
    if "optimal" in status_str.lower():
        cost_subproblem = model.getObjVal()
        total_cost = cost_vrp_travel + cost_subproblem
        return 1.0 / total_cost   # 适应度（最大化）
    else:
        return 0.00001  # 不可行解

print("--- 阶段 2: Xpress 子问题 (适应度函数) 定义完毕 ---")


# --- 阶段 3: GA 主循环 (索引映射 + 随机键编码) ---

# 1. 定义 GA 算法的参数
ga_parameters = {
    'max_num_iteration': 100,         
    'population_size': 50,            
    'parents_portion': 0.3,           
    'mutation_probability': 0.1,      
    'crossover_probability': 0.9,     
    'elit_ratio': 0.01,               
    'crossover_type': 'uniform',      
    'selection_type': 'roulette',     
    'max_iteration_without_improv': None 
}

# 2. 外循环：求解 K 和 Depot
# 假设 '1017' 是 Depot 候选站
candidate_depots_str = ['1039'] 
candidate_depots_int = [id_to_int[s] for s in candidate_depots_str if s in id_to_int]

if not candidate_depots_int:
     print("警告: 候选 Depot 站点 '1017' 缺失坐标，正在使用第一个可用站点作为 Depot。")
     candidate_depots_int = [STATIONS_INT[0]] if STATIONS_INT else [] 

if not STATIONS_INT:
    print("错误: 没有可用的站点进行优化。")
    sys.exit(1)

max_vehicles = 3 # 假设最多 10 辆车
best_overall_cost = float('inf')
best_solution = {}

start_time = time.time()

for depot_int in candidate_depots_int:
    
    # 待访问站点 = [0, 1, 2...] (不包括 depot_int)
    stations_to_visit_int = [i for i in STATIONS_INT if i != depot_int]

    for k in range(1, max_vehicles + 1):
        
        print(f"--- 正在运行 GA: Depot={int_to_id[depot_int]}, K={k} 辆车 ---")
        
        # 基因池 = [0, 1, 2...] (待访问) + [depot_int] * (k-1) (分隔符)
        genes_to_permute = stations_to_visit_int + [depot_int] * (k - 1)
        dimension = len(genes_to_permute)
        
        if dimension == 0:
            continue

        # 适应度函数封装 (使用 Random-Key 解码)
        def fitness_wrapper(chromosome):
            # 1. 解码：chromosome 是 [0,1] 间的浮点数
            permutation_indices = np.argsort(chromosome)
            
            # 2. 映射：从基因池中构建有序的整数 ID 路径
            ordered_path_list = [genes_to_permute[i] for i in permutation_indices]
            
            # 3. 解析路径 (整数索引)
            paths_by_vehicle_int = {}
            current_path = [depot_int]
            current_k = 1
            
            for station_id_int in ordered_path_list:
                if station_id_int == depot_int: # 遇到分隔符 (Depot)
                    current_path.append(depot_int)
                    paths_by_vehicle_int[current_k] = current_path
                    current_k += 1
                    current_path = [depot_int]
                else:
                    current_path.append(station_id_int)
            
            current_path.append(depot_int)
            paths_by_vehicle_int[current_k] = current_path
            
            # 4. 调用 Xpress 子问题 (传入整数路径)
            return solve_subproblem(paths_by_vehicle_int, depot_int, k)



        # 增加 Xpress 求解的超时时间，防止 GA 强制中断
        # 建议设置为 None 禁用 GA 库的超时，让 Xpress 控制求解时间
        TIMEOUT_SECONDS = 3600
        varbound = np.array([[0.0, 1.0]] * dimension)
        ga_model = ga(function=fitness_wrapper,
                      dimension=dimension,
                      variable_type='real',
                      variable_boundaries=varbound,
                      algorithm_parameters=ga_parameters,
                      function_timeout=TIMEOUT_SECONDS) # 禁用 GA 库的默认 10.0 秒超时

         

        ga_model.run()
        
        # 提取结果
        if ga_model.best_function > 0.00001:
            current_cost = 1.0 / ga_model.best_function
        else:
            current_cost = float('inf')
            
        print(f"    K={k} 完成。最优成本: {current_cost:.2f}")

        if current_cost < best_overall_cost:
            best_overall_cost = current_cost
            best_solution = {
                'Depot_Int': depot_int,
                'K_Vehicles': k,
                'Total_Cost': current_cost,
                'Optimal_Path_Chromosome (Float-Key)': ga_model.best_variable
            }

# --- 阶段 4: 结果解码与输出 ---
end_time = time.time()
print(f"\n--- 求解完成 (总耗时: {end_time - start_time:.2f} 秒) ---")

if not best_solution:
    print("未找到可行的求解方案。")
else:
    print(f"调度中心 (Depot): {int_to_id.get(best_solution.get('Depot_Int'), 'N/A')}")
    print(f"调度车辆数量 (K): {best_solution.get('K_Vehicles')}")
    print(f"最小总成本: {best_solution.get('Total_Cost'):.2f}")
    
    # 补充：最终路径解码
    final_chromosome = best_solution.get('Optimal_Path_Chromosome (Float-Key)')
    depot_int = best_solution.get('Depot_Int')
    k = best_solution.get('K_Vehicles')
    
    if final_chromosome is not None and depot_int is not None:
        stations_to_visit_int = [i for i in STATIONS_INT if i != depot_int]
        genes_to_permute = stations_to_visit_int + [depot_int] * (k - 1)
        
        permutation_indices = np.argsort(final_chromosome)
        ordered_path_list = [genes_to_permute[i] for i in permutation_indices]

        final_paths = {}
        current_path = [depot_int]
        current_k = 1
        for station_id_int in ordered_path_list:
            if station_id_int == depot_int:
                current_path.append(depot_int)
                final_paths[current_k] = current_path
                current_k += 1
                current_path = [depot_int]
            else:
                current_path.append(station_id_int)
        current_path.append(depot_int)
        final_paths[current_k] = current_path
        
        print("\n--- 最终调度路径 (原始 ID) ---")
        for k_id, path_int in final_paths.items():
            if len(path_int) > 2:
                # 解码为原始 ID
                path_str = [int_to_id[i] for i in path_int]
                print(f"车辆 {k_id} 路径: {' -> '.join(path_str)}")
```
